```python
import pandas as pd
from langdetect import detect
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# Load the dataset
data = pd.read_json("News_Category_Dataset_v3.json", lines=True)
```

```python
# Language detection
def detect_language(text):
    try:
        return detect(text)
    except:
        return "unknown"
```

```python
data.head()
```

```python
data['language'] = data['headline'].apply(detect_language)
```

```python
# Word Count and Sentence Count
data['word_count'] = data['headline'].apply(lambda x: len(word_tokenize(x)))
data['sentence_count'] = data['headline'].apply(lambda x: len(sent_tokenize(x)))

# Word-Level Tokenization
data['tokens'] = data['headline'].apply(word_tokenize)
```

```python
# Print the results
print("Language Detection Results:")
print(data['language'].value_counts())

print("\nWord Count and Sentence Count:")
print(data[['word_count', 'sentence_count']].describe())

print("\nSample Headlines with Tokens:")
print(data['tokens'].head())
```

```python
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from langdetect import detect
from collections import Counter

# Load the News category dataset
df = pd.read_json('News_Category_Dataset.json', lines=True)

# Select a sample headline from the dataset
headline = df.headline.iloc[0]

# Language detection
language = detect(headline)
print("Language Detected:", language)

# Word count
words = word_tokenize(headline)
word_count = len(words)
print("Word Count:", word_count)

# Sentence count
sentences = sent_tokenize(headline)
sentence_count = len(sentences)
print("Sentence Count:", sentence_count)

# Sentence level tokenization
tokenized_sentences = [word_tokenize(sentence) for sentence in sentences]
print("Tokenized Sentences:")
for i, tokens in enumerate(tokenized_sentences, 1):
    print(f"Sentence {i}: {tokens}")

# Word frequency count
word_frequency = Counter(words)
print("\nWord Frequency Count:")
for word, freq in word_frequency.items():
    print(f"{word}: {freq}")
```
